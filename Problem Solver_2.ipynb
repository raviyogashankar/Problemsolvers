### 10. How can you tell the difference between break and continue?

```python
# Break is used to exit the loop upon satisfying the condition
for i in range(10):
    if i == 5:
        print(i)
        break
    print("Outside the condition",i)
```

```python
# Continue is used, when a specific condition is met then the code will not execute the next steps and the next iteration will be started
list_a = [1,2,3,4,5,"Data",6,56]
for i in list_a:
    if i == "Data":
        continue
    print(i)
```

### 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
# All the three are same just the notation is different with Start ,End,Step size
print(range(10))# Only end is specified, defaultstep size is considered
print(range(0,10))# Start & end is specified, defaultstep size is considered
print(range(0,10,1))# Start,end,stepsize is specified
```
